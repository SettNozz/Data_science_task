```python
import numpy as np
from scipy.spatial import distance
import pandas as pd
from sklearn import preprocessing
```

# Input scaled data

```python
data = {"x": np.array([38.0, 45.0, 46.0, 40.0, 48.0, 38.0, 44.0, 18.0, 30.0, 40.0, 33.0, 50.0, 41.0, 28.0, 42.0]),
        "y": np.array([27.0, 17.0, 16.0, 17.0, 31.0, 39.0, 47.0, 48.0, 22.0, 18.0, 34.0, 43.0, 23.0, 39.0, 50.0]),
        "z": np.array([48.0, 33.0, 21.0, 43.0, 43.0, 16.0, 14.0, 17.0, 34.0, 36.0, 17.0, 13.0, 13.0, 40.0, 39.0])}

data_frame = pd.DataFrame(data)

scaler = preprocessing.MinMaxScaler()
data_frame[:] = scaler.fit_transform(data_frame[:])
print data_frame
```

# Functions

```python
def find_min(v, m):
    v = list(v)
    res = []
    for i in range(3):
        itm_ind = np.where(m == min(v))
        res.append(itm_ind)
        v.remove(min(v))
    print res
    
def find_max(v, m):
    v = list(v)
    res = []
    for i in range(3):
        itm_ind = np.where(m == max(v))
        res.append(itm_ind)
        v.remove(max(v))
    print res
```

# Computing distances

```python
#euclid
print "Euclidean"

dist = distance.pdist(data_frame[['x', 'y', 'z']], 'euclidean')
df_dist = pd.DataFrame(distance.squareform(dist))

print "nearest"
find_min(dist, df_dist)

print "so far"
find_max(dist, df_dist)

df_dist
```

```python
#euclid^2
print "sqrEuclidean"

dist_sq = distance.pdist(data_frame[['x', 'y', 'z']], 'sqeuclidean')
df_dist_sq = pd.DataFrame(distance.squareform(dist_sq))

print "nearest"
find_min(dist_sq, df_dist_sq)

print "so far"
find_max(dist_sq, df_dist_sq)

df_dist_sq
```

```python
#cityblock
print "cityblock"

dist_city = distance.pdist(data_frame[['x', 'y', 'z']], 'cityblock')
df_dist_city = pd.DataFrame(distance.squareform(dist_city))


print "nearest"
find_min(dist_city, df_dist_city)
v2 = list(dist_city)
print min(v2)


print "so far"
find_max(dist_city, df_dist_city)

df_dist_city
```

```python
#weighted eiclidean
dw = distance.pdist(data_frame[['x', 'y', 'z']], lambda u, v: np.sqrt(([0.5, 0.45, 0.1]*(u-v)**2).sum()))
df_dist_w = pd.DataFrame(distance.squareform(dw))


print "nearest"
find_min(dw, df_dist_w)

print "so far"
find_max(dw, df_dist_w)

df_dist_w
```
