```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

# Read and clean data

```python
#data = pd.read_csv('titanic.csv', index_col='PassengerId',usecols=['PassengerId', 'Pclass','Fare', 'Age', 'Sex'])
#data_targ = pd.read_csv('titanic.csv', index_col='PassengerId', usecols=['PassengerId','Survived'])
data = pd.read_csv('titanic.csv', index_col='PassengerId', usecols=['PassengerId', 'Pclass','Fare', 'Age', 'Sex', 'Survived'])
data = data.dropna(axis=0, how='any')
data_tart = data['Survived']
data = data.drop('Survived', 1)
spam = {'male':0, 'female':1}
data['Sex'] = data['Sex'].map(spam)
```

# Train decision tree, and fit data

```python
clf = DecisionTreeClassifier(random_state=241)
clf = clf.fit(data, data_tart)
clf.feature_importances_
```
