```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_path = '../data/illya.jpeg'
```

```python
illya_im = cv2.imread(img_path, 1)
illya_im = cv2.cvtColor(illya_im, cv2.COLOR_BGR2RGB)
gray_img = cv2.cvtColor(illya_im, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(illya_im)
```

```python
edges = cv2.Canny(gray_img, 150, 150)
plt.figure()
plt.imshow(gray_img, cmap='gray')
plt.figure()
plt.imshow(edges, cmap='gray')
```
